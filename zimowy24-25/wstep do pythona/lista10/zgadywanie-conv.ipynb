```python
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout,Embedding
```

```python
with open('brown.txt','r') as ang, open('lalka-tom-pierwszy.txt','r',encoding='utf8') as pol, open('do_wprawki.txt','r',encoding='utf8') as test:
    ANG = list({i.lower() for i in ang.read().split() if i.isalpha()})
    POL = list({i.lower() for i in pol.read().split() if i.isalpha()})
    test_set = {word.split()[1]:word.split()[0] for word in test.read().splitlines()}
    longest_word = len(max(ANG+POL+list(test_set),key = len))
```

Preprocess data, maybe find longest word, pad every word to that length, convert word to sequence of numbers

```python
ALPHABET = 'aąbcćdeęfghijklłmnńoóprsśtuwyzźżqvx'
def word_to_seq(word:str,padding:int=longest_word)->list[int]:
        #semi_converted = [ALPHABET.index(i) for i in word]
        semi_converted = []
        for i in word:
            try:
                semi_converted.append(ALPHABET.index(i))
            except:
                pass
        if padding:
            semi_converted = semi_converted+[-1]*(padding-len(semi_converted))
        return semi_converted

def seq_to_word(seq:list[int])->str:
     return ''.join(ALPHABET[i] for i in seq if i != -1)
```

```python
eng_seq = [word_to_seq(i) for i in ANG]
pol_seq = [word_to_seq(i) for i in POL]
```

```python
X = eng_seq + pol_seq
```

```python
import numpy as np
X = np.array([*X])
y = np.array([1]*len(ANG)+[0]*len(POL))
```

```python


# Define the model
model = Sequential([
    Embedding(input_dim=len(ALPHABET), output_dim=16, input_length=None),  # Dynamic input length
    Conv1D(filters=64, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Binary classification
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print the model summary
```

```python
model.fit(X, y, epochs=20, batch_size=50)
```

```python
model.summary()
```

```python
slowo = 'łamigówka'
slowo = np.array(word_to_seq(slowo)).reshape(1,24)
pred = model.predict(slowo)
if pred > 0.5:
    print('eng')
else:
    print('pol')
```
